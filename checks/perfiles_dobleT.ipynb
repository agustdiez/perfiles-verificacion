```python
# Calcular lambda_p y lambda_r para alas (según AISC 360)
lambda_pp_ala = 0.30 * np.sqrt(E / Fy)  
lambda_p_ala = 0.38 * np.sqrt(E / Fy) 
lambda_r_ala = 0.83 * np.sqrt(E / Fy)   

# Calcular lambda_p y lambda_r para alma (según AISC 360)
lambda_pp_alma = 3.06 * np.sqrt(E / Fy)  
lambda_p_alma = 3.76 * np.sqrt(E / Fy)  
lambda_r_alma = 5.70 * np.sqrt(E / Fy)


 
# Clasificar alas según bf/2tf
def clasificar_alas(bf_2tf):
    if pd.isna(bf_2tf):
        return 'N/A'
    elif bf_2tf <= lambda_pp_ala:
        return 'Muy Compacta'
    elif bf_2tf <= lambda_p_ala:
        return 'Compacta'
    elif bf_2tf <= lambda_r_ala:
        return 'No Compacta'
    else:
        return 'Esbelta'

# Clasificar alma según h/tw
def clasificar_alma(h_tw):
    if pd.isna(h_tw):
        return 'N/A'
    elif h_tw <= lambda_pp_alma:
        return 'Muy Compacta'
    elif h_tw <= lambda_p_alma:
        return 'Compacta'
    elif h_tw <= lambda_r_alma:
        return 'No Compacta'
    else:
        return 'Esbelta'

# Aplicar clasificaciones
df_simetricos_seccion['Clasificacion_Alas'] = df_simetricos_seccion['bf/2tf'].apply(clasificar_alas)
df_simetricos_seccion['Clasificacion_Alma'] = df_simetricos_seccion['h/tw'].apply(clasificar_alma)

# Clasificación general de la sección (la más crítica)
def clasificar_seccion(row):
    alas = row['Clasificacion_Alas']
    alma = row['Clasificacion_Alma']
    
    # Definir jerarquía de clasificaciones
    jerarquia = {'Muy Compacta': 0, 'Compacta': 1, 'No Compacta': 2, 'Esbelta': 3, 'N/A': 4}
    
    if alas == 'N/A' or alma == 'N/A':
        return 'N/A'
    
    # Retornar la clasificación más crítica (mayor valor en jerarquía)
    if jerarquia[alas] >= jerarquia[alma]:
        return alas
    else:
        return alma

df_simetricos_seccion['Clasificacion_Seccion'] = df_simetricos_seccion.apply(clasificar_seccion, axis=1)

# Mostrar valores de los límites
print("Límites de esbeltez calculados:")
print(f"\nAlas (bf/2tf):")
print(f"  λpp = {lambda_pp_ala:.2f} (Muy Compacta)")
print(f"  λp  = {lambda_p_ala:.2f} (Compacta)")
print(f"  λr  = {lambda_r_ala:.2f} (No Compacta)")
print(f"\nAlma (h/tw):")
print(f"  λpp = {lambda_pp_alma:.2f} (Muy Compacta)")
print(f"  λp  = {lambda_p_alma:.2f} (Compacta)")
print(f"  λr  = {lambda_r_alma:.2f} (No Compacta)")

# Mostrar resumen de clasificaciones
print("\n" + "="*50)
print("Resumen de clasificaciones:")
print("\nAlas:")
print(df_simetricos_seccion['Clasificacion_Alas'].value_counts())
print("\nAlma:")
print(df_simetricos_seccion['Clasificacion_Alma'].value_counts())
print("\nSección:")
print(df_simetricos_seccion['Clasificacion_Seccion'].value_counts())
```

```python
def compresion_simetrico(perfil_nombre, kx, ky=None, kz=None, lx=None, ly=None, lz=None, verbose=False):
    """
    Calcula capacidad a compresión de perfiles simétricos según AISC 360
    
    Parameters:
    -----------
    perfil_nombre : str
        Nombre del perfil según PERFIL_AISC
    kx, ky, kz : float
        Factores de longitud efectiva
    lx, ly, lz : float
        Longitudes no arriostradas en cm
    verbose : bool
        Imprimir detalles del cálculo
    
    Returns:
    --------
    dict : Resultados del análisis
    """
    
    # Valores por defecto
    if ky is None:
        ky = kx
    if kz is None:
        kz = kx
    if ly is None:
        ly = lx
    if lz is None:
        lz = lx
    
    
    # ========================================================================
    # PASO 1: BUSCAR PERFIL EN DATAFRAME ORIGINAL
    # ========================================================================
    try:
        perfil = simetricos[simetricos['PERFIL_AISC'] == perfil_nombre].iloc[0]
    except IndexError:
        raise ValueError(f"Perfil '{perfil_nombre}' no encontrado en la base de datos")
    
    # Propiedades geométricas del perfil
    Ag = perfil['A']      # mm²
    rx = perfil['rx']     # mm
    ry = perfil['ry']     # mm
    Ix = perfil.get('Ix', np.nan)  # mm⁴
    Iy = perfil.get('Iy', np.nan)  # mm⁴
    Zx = perfil.get('Zx', np.nan)  # mm³
    Zy = perfil.get('Zy', np.nan)  # mm³
    J = perfil.get('J', np.nan)    # mm⁴ (constante torsional de Saint-Venant)
    Cw = perfil.get('Cw', np.nan)  # mm⁶ (constante de alabeo)
    
    # Validar propiedades necesarias para pandeo torsional
    if pd.isna(J) or pd.isna(Cw):
        calcular_torsional = False
        if verbose:
            print("ADVERTENCIA: Propiedades J o Cw no disponibles. No se calculará pandeo torsional.")
    else:
        calcular_torsional = True
    
    if verbose:
        print("="*70)
        print(f"ANÁLISIS DE COMPRESIÓN - {perfil_nombre}")
        print("="*70)
        print(f"\n>>> PASO 1: Propiedades geométricas del perfil")
        print(f"    Área bruta (Ag):           {Ag:.2f} mm²")
        print(f"    Radio de giro rx:          {rx:.2f} mm")
        print(f"    Radio de giro ry:          {ry:.2f} mm")
        if calcular_torsional:
            print(f"    Constante torsional (J):   {J:.2f} mm⁴")
            print(f"    Constante de alabeo (Cw):  {Cw:.2e} mm⁶")
    
    # ========================================================================
    # PASO 2: ESBELTECES GEOMÉTRICAS
    # ========================================================================
    if verbose:
        print(f"\n>>> PASO 2: Cálculo de relaciones de esbeltez")
        print(f"    Longitud Lx:               {lx:.2f} cm = {lx*10:.0f} mm")
        print(f"    Longitud Ly:               {ly:.2f} cm = {ly*10:.0f} mm")
        print(f"    Longitud Lz:               {lz:.2f} cm = {lz*10:.0f} mm")
        print(f"    Factor Kx:                 {kx:.2f}")
        print(f"    Factor Ky:                 {ky:.2f}")
        print(f"    Factor Kz:                 {kz:.2f}")
    
    # Esbelteces geométricas
    KLx_rx = kx * lx * 10 / rx  # Convertir L de cm a mm
    KLy_ry = ky * ly * 10 / ry
    
    # Esbeltez controlante para pandeo flexional
    KL_r_flexional = max(KLx_rx, KLy_ry)
    
    if verbose:
        print(f"    KLx/rx:                    {KLx_rx:.2f}")
        print(f"    KLy/ry:                    {KLy_ry:.2f}")
        print(f"    KL/r (flexional):          {KL_r_flexional:.2f}")
    
    # ========================================================================
    # PASO 3: PANDEO FLEXIONAL (AISC 360-16 E3)
    # ========================================================================
    # Factor de esbeltez adimensional
    lambda_c_flex = (KL_r_flexional / math.pi) * math.sqrt(Fy / E)
    
    if lambda_c_flex <= 1.5:
        # Pandeo inelástico (columnas cortas/intermedias)
        Fcr_flexional = (0.658**(lambda_c_flex**2)) * Fy
        tipo_pandeo_flex = "Inelástico"
    else:
        # Pandeo elástico (columnas esbeltas)
        Fcr_flexional = (0.877 / (lambda_c_flex**2)) * Fy
        tipo_pandeo_flex = "Elástico"
    
    # Capacidad flexional
    Pn_flexional = Fcr_flexional * Ag / 1000  # kN
    
    if verbose:
        print(f"\n>>> PASO 3: Pandeo Flexional")
        print(f"    λc (esbeltez adimensional): {lambda_c_flex:.3f}")
        print(f"    Tipo de pandeo:             {tipo_pandeo_flex}")
        print(f"    Fcr (flexional):            {Fcr_flexional:.2f} MPa")
        print(f"    Pn (flexional):             {Pn_flexional:.2f} kN")
    
    # ========================================================================
    # PASO 4: PANDEO TORSIONAL PURO (AISC 360-16 E4)
    # ========================================================================
    if calcular_torsional:
        # Radio de giro polar (para secciones doblemente simétricas)
        # ro² = rx² + ry² (aproximación para el centroide)
        ro_squared = rx**2 + ry**2
        ro = math.sqrt(ro_squared)
        
        # Longitud efectiva a torsión
        Lz_mm = lz * 10  # Convertir de cm a mm
        KLz = kz * Lz_mm
        
        # Esfuerzo crítico de pandeo torsional elástico (AISC E4-4)
        # Fez = (π²·E·Cw)/(KL)² + G·J
        # Fe_torsional = Fez / (A·ro²)
        
        term1 = (math.pi**2 * E * Cw) / (KLz**2)  # Término de alabeo
        term2 = G * J                              # Término de torsión pura
        
        Fez = (term1 + term2) / (1 / (Ix + Iy))
        
        # Factor de esbeltez adimensional para torsional
        lambda_c_tors = math.sqrt(Fy / Fez)
        
        if lambda_c_tors <= 1.5:
            # Pandeo inelástico
            Fcr_torsional = (0.658**(lambda_c_tors**2)) * Fy
            tipo_pandeo_tors = "Inelástico"
        else:
            # Pandeo elástico
            Fcr_torsional = (0.877 / (lambda_c_tors**2)) * Fy
            tipo_pandeo_tors = "Elástico"
        
        # Capacidad torsional
        Pn_torsional = Fcr_torsional * Ag / 1000  # kN
        
        if verbose:
            print(f"\n>>> PASO 4: Pandeo Torsional Puro")
            print(f"    Radio de giro polar (ro):  {ro:.2f} mm")
            print(f"    KLz:                       {KLz:.0f} mm")
            print(f"    Término alabeo (π²ECw/KL²):{term1:.2e} N·mm²")
            print(f"    Término torsión (GJ):      {term2:.2e} N·mm²")
            print(f"    Fez:                       {Fez:.2f} MPa")
            print(f"    λc (torsional):            {lambda_c_tors:.3f}")
            print(f"    Tipo de pandeo:            {tipo_pandeo_tors}")
            print(f"    Fcr (torsional):           {Fcr_torsional:.2f} MPa")
            print(f"    Pn (torsional):            {Pn_torsional:.2f} kN")
    else:
        Fcr_torsional = np.nan
        Pn_torsional = np.nan
        lambda_c_tors = np.nan
        tipo_pandeo_tors = "N/A"
    
    # ========================================================================
    # PASO 5: CAPACIDAD CONTROLANTE
    # ========================================================================
    # La capacidad nominal es el MENOR valor entre pandeo flexional y torsional
    if calcular_torsional:
        Pn = min(Pn_flexional, Pn_torsional)
        Fcr = min(Fcr_flexional, Fcr_torsional)
        
        if Pn == Pn_flexional:
            modo_falla = "Pandeo Flexional"
        else:
            modo_falla = "Pandeo Torsional"
    else:
        Pn = Pn_flexional
        Fcr = Fcr_flexional
        modo_falla = "Pandeo Flexional (torsional no calculado)"
    
    # Capacidad de diseño (LRFD: φc = 0.90)
    phi_c = 0.90
    Pc = phi_c * Pn
    
    if verbose:
        print(f"\n>>> PASO 5: Capacidad Controlante")
        print(f"    Modo de falla:             {modo_falla}")
        print(f"    Fcr (controlante):         {Fcr:.2f} MPa")
        print(f"    Pn (nominal):              {Pn:.2f} kN")
        print(f"    φc Pn (diseño LRFD):       {Pc:.2f} kN")
        print("="*70)
    
    # ========================================================================
    # RETORNAR RESULTADOS
    # ========================================================================
    return {
        'perfil': perfil_nombre,
        'Ag': Ag,
        'rx': rx,
        'ry': ry,
        'ro': ro if calcular_torsional else np.nan,
        'J': J,
        'Cw': Cw,
        'KLx_rx': KLx_rx,
        'KLy_ry': KLy_ry,
        'KL_r_flexional': KL_r_flexional,
        'lambda_c_flexional': lambda_c_flex,
        'tipo_pandeo_flexional': tipo_pandeo_flex,
        'Fcr_flexional': Fcr_flexional,
        'Pn_flexional': Pn_flexional,
        'lambda_c_torsional': lambda_c_tors,
        'tipo_pandeo_torsional': tipo_pandeo_tors,
        'Fcr_torsional': Fcr_torsional,
        'Pn_torsional': Pn_torsional,
        'modo_falla': modo_falla,
        'Fcr_controlante': Fcr,
        'Pn': Pn,
        'Pc': Pc
    }

# ============================================================================
# EJEMPLO DE USO
# ============================================================================
# resultado = compresion_simetrico('W14X90', kx=1.0, lx=400, verbose=True)
# print(f"\nCapacidad de diseño: {resultado['Pc']:.2f} kN")
# print(f"Modo de falla: {resultado['modo_falla']}")
```

```python
# ============================================================================
# GENERAR DATASET DE PRUEBA - Solo K=1.0
# ============================================================================

import pandas as pd
import numpy as np
from tqdm import tqdm

print("="*70)
print("GENERANDO DATASET DE PRUEBA")
print("="*70)

# ============================================================================
# PASO 1: FILTRAR COMBINACIONES (Solo K=1.0)
# ============================================================================

# Filtrar solo combinaciones donde Kx = Ky = Kz = 1.0
df_prueba = df_combinaciones[
    (df_combinaciones['Kx'] == 1.0) & 
    (df_combinaciones['Ky'] == 1.0) & 
    (df_combinaciones['Kz'] == 1.0)
].copy()

print(f"\n>>> FILTRADO APLICADO:")
print(f"    Dataset original:      {len(df_combinaciones):,} filas")
print(f"    Dataset filtrado:      {len(df_prueba):,} filas")
print(f"    Reducción:             {(1 - len(df_prueba)/len(df_combinaciones))*100:.1f}%")

# Verificar distribución por tipo de perfil
print(f"\n>>> DISTRIBUCIÓN POR TIPO DE PERFIL:")
if 'Type' in simetricos.columns:
    # Agregar columna Type al dataset de prueba
    type_map = simetricos.set_index('PERFIL_AISC')['Type'].to_dict()
    df_prueba['Type'] = df_prueba['PERFIL_AISC'].map(type_map)
    print(df_prueba.groupby('Type')['PERFIL_AISC'].nunique())
else:
    print("    (Columna 'Type' no encontrada en dataset)")

print(f"\n>>> RESUMEN DEL DATASET DE PRUEBA:")
print(f"    Perfiles únicos:       {df_prueba['PERFIL_AISC'].nunique()}")
print(f"    Longitudes únicas:     {df_prueba['L'].nunique()}")
print(f"    Rango de L:            {df_prueba['L'].min()} - {df_prueba['L'].max()} cm")
print(f"    Valores de K:          Kx=Ky=Kz=1.0")

# ============================================================================
# PASO 2: APLICAR FUNCIÓN DE CÁLCULO
# ============================================================================

def calcular_fila(row):
    """
    Aplica la función de compresión a una fila del DataFrame
    """
    try:
        resultado = compresion_simetrico(
            perfil_nombre=row['PERFIL_AISC'],
            kx=row['Kx'],
            ky=row['Ky'],
            kz=row['Kz'],
            lx=row['L'],
            ly=row['L'],
            lz=row['L'],
            verbose=False
        )
        
        return pd.Series({
            'Pd': resultado['Pc'],
            'Pn': resultado['Pn'],
            'Fcr': resultado['Fcr_controlante'],
            'lambda_c': resultado['KL_r_flexional'],
            'modo': resultado['modo_falla'][0]
        })
        
    except Exception as e:
        return pd.Series({
            'Pd': np.nan,
            'Pn': np.nan,
            'Fcr': np.nan,
            'lambda_c': np.nan,
            'modo': 'E'
        })

# Aplicar cálculos con barra de progreso
print(f"\n{'='*70}")
print("CALCULANDO CAPACIDADES...")
print(f"{'='*70}\n")

tqdm.pandas(desc="Procesando filas")
df_resultados = df_prueba.progress_apply(calcular_fila, axis=1)

# Combinar resultados
df_final = pd.concat([df_prueba.reset_index(drop=True), df_resultados], axis=1)

print(f"\n✓ Cálculos completados")
print(f"    Filas procesadas:      {len(df_final):,}")
print(f"    Columnas totales:      {len(df_final.columns)}")

# ============================================================================
# PASO 3: VERIFICAR RESULTADOS
# ============================================================================

print(f"\n{'='*70}")
print("VERIFICACIÓN DE RESULTADOS")
print(f"{'='*70}")

# Verificar errores
if False:
    errores = df_final[df_final['modo'] == 'E']
    if len(errores) > 0:
        print(f"\n⚠️  {len(errores)} filas con errores ({len(errores)/len(df_final)*100:.2f}%)")
        print(f"    Perfiles con errores: {errores['PERFIL_AISC'].unique()[:10]}")
    else:
        print(f"\n✓  No se encontraron errores")

# Verificar valores nulos
nulos = df_final['Pd'].isna().sum()
if nulos > 0:
    print(f"\n⚠️  {nulos} valores nulos en Pd")
else:
    print(f"✓  No hay valores nulos en Pd")

# Distribución de modos de falla
print(f"\n>>> MODOS DE FALLA:")
print(df_final['modo'].value_counts())

# Estadísticas básicas
print(f"\n>>> ESTADÍSTICAS DE CAPACIDAD (Pd):")
print(df_final['Pd'].describe())

# ============================================================================
# PASO 4: OPTIMIZAR TIPOS DE DATOS
# ============================================================================

print(f"\n{'='*70}")
print("OPTIMIZANDO TIPOS DE DATOS")
print(f"{'='*70}")

print(f"\nTamaño antes: {df_final.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

# Convertir a tipos eficientes
df_final['PERFIL_AISC'] = df_final['PERFIL_AISC'].astype('category')
df_final['L'] = df_final['L'].astype('int16')
df_final['Kx'] = df_final['Kx'].astype('float16')
df_final['Ky'] = df_final['Ky'].astype('float16')
df_final['Kz'] = df_final['Kz'].astype('float16')
df_final['Pd'] = df_final['Pd'].astype('float32')
df_final['Pn'] = df_final['Pn'].astype('float32')
df_final['Fcr'] = df_final['Fcr'].astype('float32')
df_final['lambda_c'] = df_final['lambda_c'].astype('float32')
df_final['modo'] = df_final['modo'].astype('category')

if 'Type' in df_final.columns:
    df_final['Type'] = df_final['Type'].astype('category')

print(f"Tamaño después: {df_final.memory_usage(deep=True).sum() / 1024**2:.1f} MB")
print(f"Reducción: {(1 - df_final.memory_usage(deep=True).sum() / df_prueba.memory_usage(deep=True).sum())*100:.1f}%")

# ============================================================================
# PASO 5: GUARDAR ARCHIVOS
# ============================================================================

print(f"\n{'='*70}")
print("GUARDANDO ARCHIVOS")
print(f"{'='*70}\n")

# Guardar dataset de prueba
df_final.to_parquet(
    'capacidades_prueba_K1.parquet',
    engine='pyarrow',
    compression='zstd',
    compression_level=9,
    index=False
)

import os
tamaño_mb = os.path.getsize('capacidades_prueba_K1.parquet') / 1024**2
print(f"✓ Dataset guardado: capacidades_prueba_K1.parquet ({tamaño_mb:.1f} MB)")

# Guardar metadata de perfiles
columnas_meta = ['PERFIL_AISC', 'W', 'A', 'rx', 'ry', 'd', 'bf', 'tw', 'tf']
if 'Clasificacion_Alas' in simetricos.columns:
    columnas_meta.extend(['Clasificacion_Alas', 'Clasificacion_Alma'])

df_perfiles = simetricos[columnas_meta].copy()
df_perfiles.to_parquet('perfiles_metadata.parquet', index=False)
print(f"✓ Metadata guardada: perfiles_metadata.parquet")

# ============================================================================
# PASO 6: PREVIEW DE RESULTADOS
# ============================================================================

print(f"\n{'='*70}")
print("PREVIEW DE RESULTADOS")
print(f"{'='*70}\n")

# Mostrar algunos ejemplos
print(">>> PRIMERAS 10 FILAS:")
print(df_final[['PERFIL_AISC', 'L', 'Pd', 'Pn', 'lambda_c', 'modo']].head(10))

print("\n>>> ÚLTIMAS 10 FILAS:")
print(df_final[['PERFIL_AISC', 'L', 'Pd', 'Pn', 'lambda_c', 'modo']].tail(10))

# Ejemplo: Curva de un perfil específico
perfil_ejemplo = df_final['PERFIL_AISC'].iloc[0]
df_ejemplo = df_final[df_final['PERFIL_AISC'] == perfil_ejemplo]

print(f"\n>>> CURVA DEL PERFIL {perfil_ejemplo}:")
print(df_ejemplo[['L', 'Pd', 'modo']].to_string(index=False))

# ============================================================================
# RESUMEN FINAL
# ============================================================================

print(f"\n{'='*70}")
print("RESUMEN FINAL")
print(f"{'='*70}")
print(f"Dataset:           capacidades_prueba_K1.parquet")
print(f"Filas:             {len(df_final):,}")
print(f"Perfiles:          {df_final['PERFIL_AISC'].nunique()}")
print(f"Tamaño archivo:    {tamaño_mb:.1f} MB")
print(f"Condiciones:       Kx = Ky = Kz = 1.0")
print(f"Rango L:           {df_final['L'].min()} - {df_final['L'].max()} cm")
print(f"Columnas:          {', '.join(df_final.columns)}")
print(f"{'='*70}")

print(f"\n✓ ¡Dataset de prueba generado exitosamente!")
print(f"\nPróximo paso: Crear app de Shiny con este dataset\n")
```
