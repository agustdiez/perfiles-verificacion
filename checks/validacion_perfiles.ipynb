### Descripcion AISC

```python
aisc.Tipo.unique()
```

['W', 'M', 'S', 'HP', 'C', 'MC', 'L', 'WT', 'MT', 'ST', '2L', 'HSS', 'PIPE']

- Perfiles doblemente simétricos (2 ejes de simetría)

        W (Wide Flange): Perfiles I de ala ancha
        M (Miscellaneous): Perfiles I misceláneos
        S (Standard I-beam): Perfiles I estándar
        HP (Bearing Pile): Perfiles H para pilotes
        HSS (Hollow Structural Section): Perfiles tubulares rectangulares/cuadrados
        PIPE: Tubos circulares

- Perfiles con un eje de simetría

        C (Channel): Canales estándar
        MC (Miscellaneous Channel): Canales misceláneos
        WT (Structural Tee cut from W): Tee cortada de W
        MT (Structural Tee cut from M): Tee cortada de M
        ST (Structural Tee cut from S): Tee cortada de S

- Perfiles asimétricos (sin ejes de simetría)

L (Angle): Ángulos simples
2L (Double Angle): Ángulos dobles (la simetría depende de la configuración)
